```python
# Import and prepare the dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Load the Data
df = pd.read_csv('/content/train.csv')
df_test = pd.read_csv('/content/test.csv')
df.head()

# Assign the data
X = np.array(df['LotArea']).reshape(-1,1)
Y = np.array(df['SalePrice']).reshape(-1,1)
test_values = np.array(df_test['LotArea'])

# Plot the data
plt.xlim(0, 100000)
plt.xlabel('LotArea')
plt.ylabel('SalePrice')
plt.title('Lot Area vs. Sale Price')
sns.scatterplot(x=X.flatten(), y=Y.flatten())
plt.show()

# Find predicted values
model = LinearRegression()
model.fit(X, Y)

w = model.coef_
b = model.intercept_

# Plot the regression line
y_pred = model.predict(X)

plt.xlim(0, 100000)
plt.scatter(X, Y)
plt.plot(X, y_pred, color='red')
plt.xlabel('Lot Area')
plt.ylabel('Sale Price')
plt.title('Lot Area vs. Sale Price')
plt.show()

# Function to test the model with multiple values
def test_linear_regression(test_values, model):
  """
  Tests the linear regression model with a list of values.

  Args:
    test_values: A list of values to test.
    model: The trained linear regression model.
  """

  print('Testing the linear regression model with first 15 tests')
  for i in range(15):
    predicted_price = model.predict([[test_values[i]]])
    print(f'Predicted sale price for {test_values[i]:,} sqft lot: ${predicted_price[0][0]:,.2f}')

test_linear_regression(test_values, model)
```
